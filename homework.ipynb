# Univariate Analysis

```python
plt.figure(figsize=(16, 25))
for index, columns in enumerate(df.columns[1:]):
    plt.subplot(6, 2, index+1)
    plt.hist(df[columns])
    plt.title(columns)
    plt.tight_layout()
```

```python
df.groupby('STATE').size().sort_values(ascending=False)
```

```python
plt.figure(figsize=(6, 35))
for index, columns in enumerate(df.select_dtypes(include='int64').columns[1:-1]):
    plt.subplot(12, 1, index+1)
    sns.kdeplot(x=columns, data=df)
    plt.title(columns)
    plt.tight_layout()
```

```python
plt.figure(figsize=(5, 7))
for index, columns in enumerate(df.select_dtypes(include='int64').columns[1:-1]):
    plt.subplot(5, 1, index+1)
    sns.boxplot(x=columns, data=df)
    plt.tight_layout()
```
